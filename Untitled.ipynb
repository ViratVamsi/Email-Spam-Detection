```python
#same as video


def mean(numbers):
    return sum(numbers)/float(len(numbers))

def stdev(numbers):
    avg=mean(numbers)
    variance=sum([math.pow(x-avg,2) for x in numbers])/float(len(numbers)-1)
    return math.sqrt(variance)

def nb_summarize(dataset):
    summaries=[(mean(attribute),stdev(attribute)) for attribute in zip(*dataset)]
    del(summaries[-1])
    return summaries


def calculateProbability(x,mean,std):
    if(std==0):
        return 1
    exponent=math.exp(-1*math.pow(x-mean,2)/(2*math.pow(std,2)))
    return (1/(math.sqrt(2*math.pi)*std))*exponent

def ham_spam_predict(summaries,inputVector):
    probabilities={}
    for classValue,classSummaries in summaries.items():
        probabilities[classValue]=1
        for i in range(len(classSummaries)):
            mean,std=classSummaries[i]
            x=inputVector[i]
            probabilities[classValue]*=calculateProbability(x,mean,std)
    bestLabel,bestProb=None,-100.0
    for classValue,probability in probabilities.items():
        #print(classValue,'->',probability)
        if bestLabel is None or probability>bestProb:
            bestLabel=classValue
            bestprob=probability
    return bestLabel


def splitinto_labels_attributes(trainset,testset):
    #return [trainset[:-1],trainset[-1],testset[:-1],testset[-1]]
    x_train=[]
    x_test=[]
    y_train=[]
    y_test=[]
    for i in trainset:
        x_train.append(i[:-1])
        y_train.append(i[-1])
    for i in testset:
        x_test.append(i[:-1])
        y_test.append(i[-1])
    return [x_train,x_test,y_train,y_test]

def ham_spam_categorize(summaries,testSet):
    predictions=[]
    for i in range(len(testSet)):
        predictions.append(ham_spam_predict(summaries,testSet[i]))
    return predictions

def getAccuracy(testSet,predictions):
    correct=1
    for i in range(len(testSet)):
        if(testSet[i][-1]==predictions[i]):
            correct+=1
    return (correct/float(len(testSet)))*100.0


dataset=[]
with open('spambase.data','r') as spamfile:
    dataset=list(csv.reader(spamfile))
for i in range(len(dataset)):
    dataset[i]=[float(v) for v in dataset[i]]
#print(dataset)
ratio=0.75
# trainSize=int(len(dataset)*ratio)
# trainSet=[]
# testSet=list(dataset)
# while len(trainSet)<trainSize:
#     index=random.randrange(len(testSet))
#     trainSet.append(testSet.pop(index))

trainSet=[]
testSet=[]
for i in range(len(dataset)):
    if(random.random()<ratio):
        trainSet.append(dataset[i])
    else:
        testSet.append(dataset[i])

random.shuffle(trainSet)
for i in trainSet:
    random.shuffle(i[:-1])

random.shuffle(testSet)
for i in testSet:
    random.shuffle(i[:-1])

x_train,x_test,y_train,y_test=splitinto_labels_attributes(trainSet,testSet)

# for i in testSet:
#     print(i[-1])
seperated={}
for i in range(len(trainSet)):
    vector=trainSet[i]
    if(vector[-1] not in seperated):
        seperated[vector[-1]]=[]
    seperated[vector[-1]].append(vector)

summaries={}
for cv,instance in seperated.items():
    summaries[cv]=nb_summarize(instance)


predictions=ham_spam_categorize(summaries,testSet)

print(len(predictions))

accuracy=getAccuracy(testSet,predictions)
print('Accuracy: ',accuracy)
```
