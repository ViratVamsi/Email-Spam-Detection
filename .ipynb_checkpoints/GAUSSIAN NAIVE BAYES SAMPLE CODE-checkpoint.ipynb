```python
def get_mean(x_train):
    s=0
    means=[]
    l=len(x_train)
    for i in range(len(x_train[0])):
        for j in range(len(x_train)):
            s+=x_train[j][i]
        means.append(s/l)
    return means

def get_std(x_train,means):
    s=0
    std=[]
    l=len(x_train)
    for i in range(len(x_train[0])):
        for j in range(len(x_train)):
            s+=(x_train[j][i]-means[i])*(x_train[j][i]-means[i])
        std.append(math.sqrt(s/l))
    return std

def prod(l):
    pr=1
    for i in l:
        pr*=i
    return pr

# without considering any spam or ham specific attributese

def predictClasses(x_train,x_test,y_train):
    means=get_mean(x_train)
    stds=get_std(x_train,means)
    threshold=0.6
    final_probab=[]
    const1=1/(math.sqrt(2*math.pi))
    for i in range(len(x_test)):
        probab=[]
        for j in range(len(x_test[0])):
            const=1/(stds[j])
            exponent=math.exp(-1*((x_test[i][j]-means[j])*(x_test[i][j]-means[j]))/(2*stds[j]*stds[j]))
            probab.append(const*exponent)
        #print(probab)
        t=prod(probab)
#         print(t)
        if(t>=threshold):
            final_probab.append(1)
        else:
            final_probab.append(0)
    
    return final_probab


## for spam and ham specific attributes


# def get_mean(x_train,x_test):
#     sp_sum=0
#     ham_sum=0
#     spam_means=[]
#     ham_means=[]
#     sp_len=0
#     ham_len=0
#     for i in range(len(x_train[0])):
#         for j in range(len(x_train)):
#             if(x_train[j][-1]==1):
#                 sp_sum+=x_train[j][i]
#                 sp_len+=1
#             else:
#                 ham_sum+=x_train[j][i]
#                 ham_len+=1
#         #print(sp_sum,'.......',ham_sum)
#         if(sp_sum==0):
#             sp_sum=ham_sum
#         spam_means.append(sp_sum/sp_len)
#         ham_means.append(ham_sum/ham_len)
#     return [spam_means,ham_means]

# def get_std(x_train,spam_means,ham_means,x_test):
#     sp_sum=0
#     ham_sum=0
#     spam_stds=[]
#     ham_stds=[]
#     sp_len=0
#     ham_len=0
#     for i in range(len(x_train[0])):
#         for j in range(len(x_train)):
#             if(x_train[j][-1]==1):
#                 sp_sum+=(x_train[j][i]-spam_means[i])*(x_train[j][i]-spam_means[i])
#                 sp_len+=1
#             else:
#                 ham_sum+=(x_train[j][i]-ham_means[i])*(x_train[j][i]-ham_means[i])
#                 ham_len+=1
            
#         spam_stds.append(math.sqrt(sp_sum/sp_len))
#         ham_stds.append(math.sqrt(ham_sum/ham_len))
#     return [spam_stds,ham_stds]

# def predictClasses(x_train,x_test,y_train):
#     spam_means,ham_means=get_mean(x_train,x_test)
#     spam_stds,ham_stds=get_std(x_train,spam_means,ham_means,x_test)
#     #print(spam_stds,ham_stds)
#     threshold=0.6
#     final_probab=[]
#     const1=1/(math.sqrt(2*math.pi))
#     for i in range(len(y_train)):
#         spam_probab=[]
#         ham_probab=[]
#         for j in range(len(y_train[0])):
#             sp_const=1/(spam_stds[j])
#             sp_exponent=math.exp(-1*((y_train[i][j]-spam_means[j])*(y_train[i][j]-spam_means[j]))/(2*spam_stds[j]*spam_stds[j]))
#             spam_probab.append(sp_const*sp_exponent)
            
#             ham_const=1/(ham_stds[j])
#             ham_exponent=math.exp(-1*((y_train[i][j]-ham_means[j])*(y_train[i][j]-ham_means[j]))/(2*ham_stds[j]*ham_stds[j]))
#             ham_probab.append(ham_const*ham_exponent)
# #         print(probab)
# #         t=prod(probab)
# #         print(t)
#         if(prod(spam_probab)>prod(ham_probab)):
#             final_probab.append(1)
#         else:
#             final_probab.append(0)
    
#     return final_probab
```

```python
def split_traintest_data(trainset,testset,split):
    with open('spambase.data','r') as spamfile:
        lines=list(csv.reader(spamfile))
        for i in range(len(lines)):
            for j in range(57):
                lines[i][j]=float(lines[i][j])
            lines[i][-1]=int(lines[i][-1])
        for i in lines:
            if(random.random()<split):
                trainset.append(i)
            else:
                testset.append(i)
        random.shuffle(trainset)
        for i in trainset:
            random.shuffle(i[:-1])
        
        random.shuffle(testset)
        for i in testset:
            random.shuffle(i[:-1])
def splitinto_labels_attributes(trainset,testset):
    #return [trainset[:-1],trainset[-1],testset[:-1],testset[-1]]
    x_train=[]
    x_test=[]
    y_train=[]
    y_test=[]
    for i in trainset:
        x_train.append(i[:-1])
        y_train.append(i[-1])
    for i in testset:
        x_test.append(i[:-1])
        y_test.append(i[-1])
    return [x_train,y_train,x_test,y_test]


def calculate_accuracy(y_test,model_predicted):
    if(len(y_test)!=len(model_predicted)):
        print('Calculation Error, length not same')
    error=0
    for i in range(len(y_test)):
        if(y_test[i]==model_predicted[i]):
            error+=1
    error_percent=((error)/len(y_test))*100
    return error_percent

split=0.75
trainset=[]
testset=[]
split_traintest_data(trainset,testset,split)
#print(len(trainset),len(testset))  #3089 1517  approx around these values since random split
x_train,y_train,x_test,y_test=splitinto_labels_attributes(trainset,testset)

model_predicted=predictClasses(x_train,x_test,y_train)

# print(len(y_test))
# print(len(model_predicted))

# for i in range(len(y_test)):
#     print(y_test[i],'->',model_predicted[i])

accuracy=calculate_accuracy(y_test,model_predicted)
print('accuracy is ',accuracy,'%')
```
