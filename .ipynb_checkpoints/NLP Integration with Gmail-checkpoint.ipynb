```python
import csv
import random
import math
def split_traintest_data(trainset,testset,split):
    with open('spambase.data','r') as spamfile:
        lines=list(csv.reader(spamfile))
        for i in range(len(lines)):
            for j in range(57):
                lines[i][j]=float(lines[i][j])
            lines[i][-1]=int(lines[i][-1])
        for i in lines:
            if(random.random()<split):
                trainset.append(i)
            else:
                testset.append(i)
        random.shuffle(trainset)
        for i in trainset:
            random.shuffle(i[:-1])
        
        random.shuffle(testset)
        for i in testset:
            random.shuffle(i[:-1])
def splitinto_labels_attributes(trainset,testset):
    #return [trainset[:-1],trainset[-1],testset[:-1],testset[-1]]
    x_train=[]
    x_test=[]
    y_train=[]
    y_test=[]
    for i in trainset:
        x_train.append(i[:-1])
        x_test.append(i[-1])
    for i in testset:
        y_train.append(i[:-1])
        y_test.append(i[-1])
    return [x_train,x_test,y_train,y_test]

def euclid_dist(sample_ytrain,sample_xtrain):
    dist=0
    for i in range(len(sample_ytrain)):
        diff=(sample_ytrain[i]-sample_xtrain[i])
        dist+=diff*diff
    return math.sqrt(dist)

def nearest_neighbours(distances,k):
    knn=[]
#     m=0
#     n=0
#     o=0
#     for i in range(len(distances)):
#         if(i[m][0]>distances[i][0]):
#             m=i
#     knn.append([distances[m][0],distances[m][1]])
#     for i in range(len(distances)):
#         if(i==m):
#             continue
#         if(i[n][0]>distances[i][0]):
#             n=i
#     knn.append([distances[n][0],distances[n][1]])
#     for i in range(len(distances)):
#         if(i==m or i==n):
#             continue
#         if(i[o][0]>distances[i][0]):
#             0=i
#     knn.append([distances[o][0],distances[o][1]])
    distances.sort(key=lambda x:x[0])
    for i in range(k):
        knn.append(distances[i])
    return knn
    
def knn_predictlabel(knn):
    onecount=0
    for i in knn:
        if(i[1]==1):
            onecount+=1
    if(onecount<=len(knn)-onecount):
        return 0
    else:
        return 1

def predictClasses(x_train,x_test,y_train,k):
    predicted_labels=[]
    for i in range(len(y_train)):
        distances=[]
        for j in range(len(x_train)):
            distances.append([euclid_dist(y_train[i],x_train[j]),x_test[j]])
        knn=nearest_neighbours(distances,k)
        #print(knn)
        predicted_labels.append(knn_predictlabel(knn))
    return predicted_labels

def calculate_accuracy(y_test,model_predicted):
    if(len(y_test)!=len(model_predicted)):
        print('Calculation Error, length not same')
    error=0
    for i in range(len(y_test)):
        if(y_test[i]==model_predicted[i]):
            error+=1
    error_percent=((error)/len(y_test))*100
    return error_percent

split=0.67
k=3
trainset=[]
testset=[]
split_traintest_data(trainset,testset,split)
#print(len(trainset),len(testset))  #3089 1517  approx around these values since random split
x_train,x_test,y_train,y_test=splitinto_labels_attributes(trainset,testset)
# for i in testset:
#     print(i[-1])
#print(len(x_train),len(x_test),len(y_train),len(y_test))

model_predicted=predictClasses(x_train,x_test,y_train,k)
# print(len(y_test))
# print(len(model_predicted))


# for i in range(len(y_test)):
#     print(y_test[i],'->',model_predicted[i])

accuracy=calculate_accuracy(y_test,model_predicted)
print('accuracy is ',accuracy,'%')
```

### Integration testing for Gaussian NaiveBayes

```python
print("\nNaive Bayes Classifier for concept learning problem")
import csv
import random
import math
import operator
def safe_div(x,y):
    if y == 0:
        return 0
    return x / y
	
def loadCsv(filename):
	lines = csv.reader(open(filename))
	dataset = list(lines)
	for i in range(len(dataset)):
		dataset[i] = [float(x) for x in dataset[i]]
	return dataset
 
def splitDataset(data,ratio):
    dataset=[]
    with open(data,'r') as spamfile:
        dataset=list(csv.reader(spamfile))
    for i in range(len(dataset)):
        for j in range(57):
            dataset[i][j]=float(dataset[i][j])
        dataset[i][-1]=int(dataset[i][-1])
    trainSize=int(len(dataset)*ratio)
    trainSet=[]
    copy=list(dataset)
    while len(trainSet)<trainSize:
        index=random.randrange(len(copy))
        trainSet.append(copy.pop(index))
    random.shuffle(trainSet)
    for i in trainSet:
        random.shuffle(i[:-1])

    random.shuffle(copy)
    for i in copy:
        random.shuffle(i[:-1])
    return [trainSet,copy]

def separateByClass(dataset):
	separated = {}
	for i in range(len(dataset)):
		vector = dataset[i]
		if (vector[-1] not in separated):
			separated[vector[-1]] = []
		separated[vector[-1]].append(vector)
	return separated
 
def mean(numbers):
	return safe_div(sum(numbers),float(len(numbers)))
	
def stdev(numbers):
	avg = mean(numbers)		
	variance = safe_div(sum([pow(x-avg,2) for x in numbers]),float(len(numbers)-1))
	return math.sqrt(variance)
 
def summarize(dataset):
	summaries = [(mean(attribute), stdev(attribute)) for attribute in zip(*dataset)]
	del summaries[-1]
	return summaries
 
def summarizeByClass(dataset):
	separated = separateByClass(dataset)
	summaries = {}	
	for classValue, instances in separated.items():		
		summaries[classValue] = summarize(instances)
	return summaries
 
# def calculateProbability(x, mean, stdev):	
# 	exponent = math.exp(-safe_div(math.pow(x-mean,2),(2*math.pow(stdev,2))))
# 	final = safe_div(1 , (math.sqrt(2*math.pi) * stdev)) * exponent
# 	return final

def calculateProbability(x,mean,std):
    exponent=math.exp(-1*math.pow(x-mean,2)/(0.000001+2*math.pow(std,2)))
    return (1/(0.000001+math.sqrt(2*math.pi)*std))*exponent
 
def calculateClassProbabilities(summaries, inputVector):
	probabilities = {}
	for classValue, classSummaries in summaries.items():
		probabilities[classValue] = 1
		for i in range(len(classSummaries)):
			mean, stdev = classSummaries[i]			
			x = inputVector[i]
			probabilities[classValue] *= calculateProbability(x, mean, stdev)
	return probabilities
			
def predict(summaries, inputVector):
	probabilities = calculateClassProbabilities(summaries, inputVector)
	bestLabel, bestProb = None, -1
	for classValue, probability in probabilities.items():
		if bestLabel is None or probability > bestProb:
			bestProb = probability
			bestLabel = classValue
	return bestLabel
 
def getPredictions(summaries, testSet):
	predictions = []
	for i in range(len(testSet)):
		result = predict(summaries, testSet[i])
		predictions.append(result)
	return predictions
 
def getAccuracy(testSet, predictions):
	correct = 0
	for i in range(len(testSet)):
		if testSet[i][-1] == predictions[i]:
			correct += 1
	accuracy = safe_div(correct,float(len(testSet))) * 100.0
	return accuracy



def pred():
    trainSet,testSet=splitDataset('spambase.data',0.67)
    frequency_list.append(0)
    testSet.append(frequency_list)
    
    summaries=summarizeByClass(trainSet)
    predictions=getPredictions(summaries,testSet)
    accuracy=getAccuracy(testSet,predictions)
    for i in range(len(testSet)):
        print(testSet[i][-1],'->',predictions[i])
    print('Accuracy: ',accuracy)
    
    
#     trainSet,testSet=splitDataset('spambase.data',0.67)
#     summaries=summarizeByClass(trainSet)
#     frequency_list.append(0)
#     sample_test=[frequency_list]
#     print(sample_test)
#     predictions=getPredictions(summaries,sample_test)
#     for i in predictions:
#         if(i==0):
#             print('Your Mail is Ham')
#         else:
#             print('Alert!!! Your Mail is spam')


#     accuracy=getAccuracy(testSet,predictions)
#     for i in range(len(testSet)):
#         print(testSet[i][-1],'->',predictions[i])
#     print('Accuracy: ',accuracy)
pred()
```
